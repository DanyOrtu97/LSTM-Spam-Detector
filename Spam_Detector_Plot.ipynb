```
import matplotlib.pyplot as plt
import numpy as np
```

**PLOT Grafici accuratezza RNN1**



Batch size 16

```
learning_rate = [0.0005, 0.005, 0.001]
accuracy_MSE_RMSprop = [98.43, 81.1, 98.4]
accuracy_BCE_RMSprop = [98.33, 98.31, 98.47]
accuracy_MSE_Adam = [98.07, 91.22, 98.33]
accuracy_BCE_Adam = [99.11, 99.24, 99.26]

accuracy = accuracy_MSE_RMSprop + accuracy_BCE_RMSprop + accuracy_MSE_Adam + accuracy_BCE_Adam

plt.figure(figsize=(25,10))
plt.plot(learning_rate, accuracy_MSE_RMSprop, marker='o', linestyle='dashed', linewidth=2, markersize=6, color="red", label='MSE_RMSprop')
plt.plot(learning_rate, accuracy_BCE_RMSprop, marker='*', linestyle='dashed', linewidth=2, markersize=6, color="blue", label='BCE_RMSprop')
plt.plot(learning_rate, accuracy_MSE_Adam, marker='+', linestyle='dashed', linewidth=2, markersize=6, color="green", label='MSE_Adam')
plt.plot(learning_rate, accuracy_BCE_Adam, marker='x', linestyle='dashed', linewidth=2, markersize=6, color="yellow", label='BCE_Adam')

plt.grid(color='b', linestyle='-', linewidth=0.1)
plt.xticks(learning_rate, rotation=0)
plt.yticks(accuracy, rotation=0)
plt.ylabel("accuracy train (%)", fontsize=13)
plt.xlabel("learning rate", fontsize=13)
plt.legend(fancybox=True, borderpad=1, loc="upper right", prop={'size': 15})




```
