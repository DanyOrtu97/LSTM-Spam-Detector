```python
import matplotlib.pyplot as plt
import numpy as np
```

**PLOT Grafici accuratezza**

```python
RNN1 = np.load("RNN1W2V.npy")
```

```python
learning_rate = [0.0005, 0.005, 0.001]
batch_sizes = [16,32,64,128]
```

```python
lr1 = RNN1[np.where(RNN1[:,3]=="0,0005")]
```

```python
RMS_MSE = lr1[np.logical_and(lr1[:,1] == "MSE" , lr1[:,2] == "RMSprop")]
RMS_BCE = lr1[np.logical_and(lr1[:,1] == "BCE" , lr1[:,2] == "RMSprop")]
Adam_MSE = lr1[np.logical_and(lr1[:,1] == "MSE" , lr1[:,2] == "Adam")]
Adam_BCE = lr1[np.logical_and(lr1[:,1] == "BCE" , lr1[:,2] == "Adam")]
```

```python
fig = plt.figure(figsize=(25,10))
plt.title("RNN1glove - Learning Rate = 0,0005",fontdict={'fontsize': 20, 'fontweight': 'bold'})
plt.plot(RMS_MSE[:,0], RMS_MSE[:,5], marker='o', linestyle='dashed', linewidth=2, markersize=6, color="red", label='MSE_RMSprop')
plt.plot(RMS_BCE[:,0], RMS_BCE[:,5], marker='*', linestyle='dashed', linewidth=2, markersize=6, color="blue", label='BCE_RMSprop')
plt.plot(Adam_MSE[:,0], Adam_MSE[:,5], marker='+', linestyle='dashed', linewidth=2, markersize=6, color="green", label='MSE_Adam')
plt.plot(Adam_BCE[:,0], Adam_BCE[:,5], marker='x', linestyle='dashed', linewidth=2, markersize=6, color="orange", label='BCE_Adam')

plt.grid(color='b', linestyle='-', linewidth=0.1)
plt.ylabel("Accuracy Test (%)", fontsize=13)
plt.xlabel("Batch Size", fontsize=13)

plt.legend(fancybox=True, borderpad=1, loc="upper left", prop={'size': 15})
```
